### Сбор данных (.csv файлы вложенных папок в data_dir) в единый df

```python
from itertools import chain
all_files = [tuple_[2] for tuple_ in  list(os.walk(data_dir))[1:]]
all_files = [files for files in all_files if any(file.endswith('.csv') for file in files)]
all_files = list(chain.from_iterable(all_files))
all_new_data = list(os.walk(data_dir))[1:]
all_new_data = [(root, dirs, files) for root, dirs, files in all_new_data if any(file in all_files for file in files)]
all_new_data
```

```python
amount_of_files = sum([len(files) for files in [files[2] for files in all_new_data]])
amount_of_files
```

```python
all_csv_paths = []

for i in range(len(all_new_data)):
    for j in range(len(all_new_data[i][2])):
        all_csv_paths.append((all_new_data[i][0], all_new_data[i][2][j]))

errors_list = [] # сюда положим те файлы с данными, которые не получается обработать
data_dir, csv_file = all_csv_paths[0]
df = tools.create_correct_df(data_dir, csv_file)
for data_dir, csv_file in all_csv_paths[1:]:
    try:
        df = tools.concat_next_csv(df, data_dir, csv_file)
    except UnicodeDecodeError:
        errors_list.append((data_dir, csv_file))
df
```

```python
print(len(errors_list))
errors_list
```

### Попробуем обработать файлы с данными из errors_list с другой кодировкой

```python
new_errors_list = []
for data_dir, csv_file in errors_list:
    try:
        df = tools.concat_next_csv(df, data_dir, csv_file, "ISO-8859-1")
    except Exception:
        new_errors_list.append((data_dir, csv_file))
        continue
errors_list = new_errors_list
print(len(errors_list))
errors_list
```
