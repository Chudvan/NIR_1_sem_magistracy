### Импорт библиотек

```python
import os
import sys

project_dir_path = os.environ["NIR_1_sem_magistracy"]
data_dir = 'Data'
data_dir = os.path.join(project_dir_path, data_dir)
print('data_dir:', data_dir)

lib_dir = 'lib'
lib_dir_path = os.path.join(project_dir_path, lib_dir)
print('lib_dir_path:', lib_dir_path)
sys.path.append(lib_dir_path)
import nn_train.tools as tools
```

### Сбор данных (ТОЛЬКО .csv файлы из папки data_dir /* без вложенных папок */) в единый df

```python
csv_files = os.listdir(data_dir)
csv_files = [csv_file for csv_file in csv_files if csv_file.endswith('.csv')]
csv_files
```

```python
df = tools.create_correct_df(data_dir, csv_files[0])
for csv_file in csv_files[1:]:
    df = tools.concat_next_csv(df, data_dir, csv_file)
df
```

### Запись в БД 'Aleksashina_Керов_Андрей.db'

```python
name_db = 'Aleksashina_Керов_Андрей'
db_file = f"{name_db}.db"
db_dir = 'databases'
db_path = os.path.join(data_dir, db_dir, db_file)
print(db_path)
# нужно удалить файл db_file = f"{name_db}.db" перед запуском
try:
    os.remove(db_path)
except FileNotFoundError:
    print(f"No such file or directory: '{db_path}'")
```

```python
tools.save_to_db(db_path, name_db, df)
```
