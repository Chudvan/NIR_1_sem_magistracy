```python
import pandas as pd
```

```python
import os
import sqlite3
```

```python
data_dir = 'Data'
name_db = 'Aleksashina_Керов_Андрей'
db_file = f"{name_db}.db"
db_path = os.path.join(data_dir, db_file)
```

```python
connection = sqlite3.connect(db_path)
df = pd.read_sql(f'select * from {name_db}', con=connection)
df
```

```python
df['Arousal'].apply(lambda x: float(x))
```

```python
seven_fields = [
    'Neutral', 
    'Happy', 
    'Sad', 
    'Angry', 
    'Surprised', 
    'Scared', 
    'Disgusted'
]

for field in seven_fields:
    df[field] = df[field].apply(lambda x: float(x))
```

```python
seven_vector = df[seven_fields]
```

```python
seven_vector
```

```python
seven_vector.index = list(df['Index_'])
seven_vector
```

```python
pa_fields =     [
    'Valence',
    'Arousal'
]

for field in pa_fields:
    df[field] = df[field].apply(lambda x: float(x))
```

```python
pa_vector = df[pa_fields]
```

```python
pa_vector
```

```python
pa_vector.index = list(df['Index_'])
pa_vector
```

```python
import numpy as np
```

```python
x = []
for i in range(len(pa_vector)):
    x.append((pa_vector['Valence'][i], pa_vector['Arousal'][i])) 
    
x = np.array(x)
```

```python
x.shape
```

```python
y = []
for i in range(len(pa_vector)):
    y.append(tuple(seven_vector[col][i] for col in seven_vector.columns)) 
y = np.array(y)
```

```python
y.shape
```

```python
from keras.models import Model
from keras.layers import Input, Dense
```

```python
a = Input(shape=(2,))
first = Dense(5)(a)
second = Dense(6)(first)
third = Dense(7)(second)
b = Dense(7)(third)
model = Model(inputs=a, outputs=b)
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=x, y=y, epochs=10)
```

```python
from tensorflow.python.keras.utils.vis_utils import model_to_dot
from tensorflow.python.keras.utils.vis_utils import plot_model
```

```python
class NeuralNetwork:
    def __init__(self, x, y, model):
        self.df_x = x
        self.df_y = y
        self.model = model
        self.from_ = 0
        self.to = 1
    
    def get_df_xy(self, from_ = None, to = None, xy = None):
        if xy is None:
            xy = 'x'
        
        if from_ is None:
            from_ = self.from_
        else:
            self.from_ = from_
            
        if to is None:
            to = self.to
        else:
            self.to = to
        
        if xy.lower() == 'x':   
            return self.df_x[from_:to]
        elif xy.lower() == 'y':   
            return self.df_y[from_:to]
    
    def predict(self):
        df_x = self.get_df_xy(self.from_, self.to)
        df_res = pd.DataFrame(self.model.predict(df_x.values))
        df_res.columns = self.df_y.columns
        df_res.index = df_x.index
        return df_res
```
