```python
import pandas as pd
import numpy as np
```

```python
# Код со stackoverflow

from IPython.display import display_html
from itertools import chain,cycle


def display_side_by_side(*args, titles=cycle([''])):
    html_str=''
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        html_str+='<td style="vertical-align:top">'
        html_str+=f'<h2>{title}</h2>'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+='</td>'
    display_html(html_str,raw=True)
```

```python
df1 = pd.DataFrame(np.arange(12).reshape((3,4)),columns=['A','B','C','D',])
df2 = pd.DataFrame(np.arange(16).reshape((4,4)),columns=['A','B','C','D',])
display_side_by_side(df1,df2,df1, titles=['Foo','Foo Bar']) #we left 3rd empty...
```

```python
# Для моей задачи

from IPython.display import display_html
from itertools import chain,cycle


def display_dfs(*args, titles=cycle(['']), mode='column'):
    html_str=''
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        cur_html_str = ''
        cur_html_str+='<td style="vertical-align:top">'
        cur_html_str+=f'<h2>{title}</h2>'
        cur_html_str+=df.to_html().replace('table','table style="display:inline"')
        cur_html_str+='</td>'
        if mode == 'column':
            cur_html_str = '<tr>' + cur_html_str + '</tr>'
        elif mode == 'row':
            pass
        else:
            raise Exception(f'Unknown mode: {mode}')
        html_str += cur_html_str
    display_html(html_str,raw=True)
```

```python
df1 = pd.DataFrame(np.arange(12).reshape((3,4)),columns=['A','B','C','D',])
df2 = pd.DataFrame(np.arange(16).reshape((4,4)),columns=['A','B','C','D',])
```

```python
display_dfs(df1,df2,df1, titles=['Foo','Foo Bar']) #we left 3rd empty...
```

```python
display_dfs(df1,df2,df1, titles=['Foo','Foo Bar'], mode='row') #we left 3rd empty...
```

```python
display_dfs(df1,df2,df1, titles=['Foo','Foo Bar'], mode='asd') #we left 3rd empty...
```
